Its time to remove outliers in the data set

```python
#calculating upper and lower limits
upper_limit=pd.Order_Quantity.quantile(0.25)
lower_limits=pd.Order_Quantity.quantile(0.75)
#compute Interquatile range
IQR=upper_limit-lower_limits

Upper_limits=upper_limit-1.5*IQR
Lower_limits=lower_limits+1.5*IQR

pd.loc[(pd.Order_Quantity<Lower_limits)|(pd.Order_Quantity>Upper_limits)]
```

```python
pd.loc[(pd.Order_Quantity>Lower_limits)&(pd.Order_Quantity>Upper_limits)]
```

```python
X=pd.drop(columns=("Revenue"),axis=1)
Y=pd['Revenue']
```

Importing libraries for data modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.model_selection import train_test_split
```

Splitting data into train and test

```python
# Train and test
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
#Rows and columns
X.shape,X_train.shape,X_test.shape
```

```python
#creating a model with linear reggression
model=LinearRegression()
model.fit(X_train,Y_train)
```

Model evaluation

```python
model.score(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
y_pred_train=model.predict(X_train)
y_pred_test=model.predict(X_test)
```

```python
print("Rsq score error:\n{}".format(r2_score(Y_train,y_pred_train)))
```

```python

print("Rsq score error:\n{}".format(r2_score(Y_test,y_pred_test)))
```

```python
print("Mean absolute error:\n{}".format(mean_absolute_error(Y_train,y_pred_train)))
```

```python

print("Mean absolute error:\n{}".format(mean_absolute_error(Y_test,y_pred_test)))
```

```python
print("Mean square error:\n {}".format(mean_squared_error(Y_train,y_pred_train)))
```

```python

print("Mean square error:\n {}".format(mean_squared_error(Y_test,y_pred_test)))
```
